# Libs Necessárias

```python
# Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

# Libs para anaálises gráficas
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

# Lib para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
# Lendo os dados
Base_Dados = pd.read_excel('Vase_004 - Magalu.xlsx')
```

```python
# Verificando cabeçalhos 
Base_Dados.head()
```

```python
#representa o número de linhas e colunas
Base_Dados.shape
```

```python
#Obter informações sobre os tipos de dados
Base_Dados.info()
```

```python
# Descrição estatística da base de dados
Base_Dados.describe()
```

```python
# Series Temporais
Dados = Base_Dados.set_index('Data')
Dados.head()
```

```python
plt.style.use('seaborn-darkgrid')
plt.figure( figsize=(16, 5) )
plt.title('Análise das ações da magalu - Fechamento', fontsize=15, loc='left')
plt.plot( Dados.index, Dados['Fechamento'] )

plt.xlabel('Período da Coração')
plt.ylabel('Valor da Ação (R$)');
```

```python
#Ultimo dados 
Dados.tail()
```

```python
Media_Movel = Dados['Fechamento'].rolling(5).mean()#seleciona de 5 em 5 tirando a meia movel de 5 dias
Media_Tendencia = Dados['Fechamento'].rolling(30).mean()#agrupada durante 30 dias

plt.style.use('seaborn-darkgrid')
plt.figure( figsize=(16, 5) )
plt.title('Análise das ações da magalu - Fechamento', fontsize=15, loc='left')

plt.plot( Dados.index, Dados['Fechamento'] )
plt.plot( Dados.index, Media_Movel )
plt.plot( Dados.index, Media_Tendencia )

plt.xlabel('Período da Coração')
plt.ylabel('Valor da Ação (R$)');
```

```python
sns.boxplot( data=Dados, x='Fechamento')
```

```python
# Boxplot Mensal 

Base_Dados['Mes'] = Base_Dados['Data'].dt.month

plt.figure( figsize=(16, 5) )
sns.boxplot( data=Base_Dados, x='Mes', y='Fechamento' );
```

```python
Base_Dados.groupby(['Mes']).describe()['Fechamento']
```
